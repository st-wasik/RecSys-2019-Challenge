```python
import pandas as pd

meta_data = pd.read_csv("splitterOut/item_metadata.csv")
meta_data
```

```python
data_set = pd.read_csv("sortedTrain.csv")
```

```python
def fastSplitting(df):#stack overflow ❤ 
    df2 = df.loc[df["properties"].str.contains("Good Rating")]
    s = df2["properties"].str.split('|', expand=True).stack()
    i = s.index.get_level_values(0)
    df3 = df.loc[i].copy()
    df3["properties"] = s.values
    return df3

traits = fastSplitting(meta_data).groupby('properties').count().sort_values(by=['item_id'],ascending=False).nlargest(70, 'item_id')
```

```python
traits
```

```python
import numpy as np
traits2 = traits.reset_index()
traits2 = traits2['properties'].str.replace(' ', '_')
```

```python
traits2.values
```

```python
meta_data['properties'] = meta_data['properties'].str.replace(' ','_')
    
for i in traits.values:
    meta_data[i] = np.where( meta_data.properties.str.contains('|' + i + '|', regex=False), 1, 0)
```
